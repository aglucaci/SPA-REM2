```python
import os
import sys
import pandas as pd
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
#import os
from tqdm import tqdm
import numpy as np
import altair as alt
from vega_datasets import data
alt.data_transformers.enable('default', max_rows=None)
```

```python
df_tmalign = pd.read_csv(os.path.join("..", 
                         "tables", 
                         'TM_Align_Results.csv'), index_col=0)

df_tmalign
```

```python
df_tn93 = pd.read_csv(os.path.join("..", 
                      "results", "TN93",
                      'tn93_rem2.dst'))

df_tn93
```

```python
df_tmalign["Short_Name of Chain_1:"] = ""
df_tmalign["Short_Name of Chain_2:"] = ""

def parseName(name):
    parsed_name = name
    parsed_name = parsed_name.split("_")
    if "Node" in parsed_name[0]:
        parsed_name = parsed_name[0]
    else:
        parsed_name = "_".join(parsed_name[0:3])
    #end if
    return parsed_name
#end method

for index, row in tqdm(df_tmalign.iterrows()):
    id1 = row["Name of Chain_1:"]
    id2 = row["Name of Chain_2:"]
    
    # Parse names
    id1_parsed = parseName(id1)
    id2_parsed = parseName(id2)
    
    # dfmi.loc[:, ('one', 'second')]
    
    #df_tmalign.loc[:, ("Short_Name of Chain_1:", index)] = id1_parsed
    #df_tmalign.loc[:, ("Short_Name of Chain_2:", index)] = id2_parsed
    
    df_tmalign["Short_Name of Chain_1:"][index] = id1_parsed
    df_tmalign["Short_Name of Chain_2:"][index] = id2_parsed
#end for


```

```python
df_tmalign
```

```python
df_tmalign["TN93"] = ""

for index, row in tqdm(df_tn93.iterrows()):
    id1_parsed = parseName(row["ID1"])
    id2_parsed = parseName(row["ID2"])
    distance =   row["Distance"]
    
    df = df_tmalign[df_tmalign["Short_Name of Chain_2:"] == id1_parsed]
    df2 = df[df["Short_Name of Chain_1:"] == id2_parsed]
    df_tmalign["TN93"][df2.index] = distance
    
    df = df_tmalign[df_tmalign["Short_Name of Chain_1:"] == id1_parsed]
    df2 = df[df["Short_Name of Chain_2:"] == id2_parsed]
    df_tmalign["TN93"][df2.index] = distance
        
    #break
    #print(df2.index[0])
    #break
```

```python

# We will add things to the 'results_dict' dictionary
# We need to go two levels down for it

for k in tqdm(results_dict.keys()):
    for k2 in results_dict[k].keys():
        pair = [k, k2]
        
        for k_t in results_dict_tn93.keys():
            for k_t2 in results_dict_tn93.keys():
                pair_t = [k_t, k_t2]
                
                if set(pair) == set(pair_t): # Match
                    #print('match')
                    try:
                        results_dict[k][k2].update({"TN93_GENETIC_DISTANCE": results_dict_tn93[k_t][k_t2]['TN93']})
                    except:
                        results_dict[k][k2].update({"TN93_GENETIC_DISTANCE": results_dict_tn93[k_t2][k_t]['TN93']})
                    #end try
                        
                    if results_dict[k][k2].get("TN93_GENETIC_DISTANCE", np.nan) == np.nan:
                        try:
                            results_dict[k2][k].update({"TN93_GENETIC_DISTANCE": results_dict_tn93[k_t][k_t2]['TN93']})
                        except:
                            results_dict[k2][k].update({"TN93_GENETIC_DISTANCE": results_dict_tn93[k_t2][k_t]['TN93']})
                        #endtry
                    #end if
                #end if
            #end for
        #end for
    #end for
#end for
        
```

```python
#results_dict
## dict to df
## df to csv

formatted_dict = {}
count = 1

for k in results_dict.keys():
    for k2 in results_dict[k].keys():
        temp_dict = {}
        for item in results_dict[k][k2].keys():
            temp_dict.update({item : results_dict[k][k2][item]})
        #end for
        
        formatted_dict[count] = {"ID1": k,
                                 "ID2": k2}
        formatted_dict[count].update(temp_dict)                           
        
        count +=1 
    #end for
    
#end for
```
